<h1 style="color:skyblue;text-align:center;">Face Recognition Model Using FaceNet,MTCNN,SVM</h1>

```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from mtcnn import MTCNN
from keras_facenet import FaceNet
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import pickle
import logging
```

```python
# Set up logging
logging.basicConfig(level=logging.INFO)
```

```python
class FACELOADING:
    def __init__(self, directory):
        self.directory = directory
        self.target_size = (160, 160)
        self.X = []
        self.Y = []
        self.detector = MTCNN()
    
    def extract_faces(self, filename):
        img = cv2.imread(filename)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        try:
            x, y, w, h = self.detector.detect_faces(img)[0]['box']
            x, y = abs(x), abs(y)
            face = img[y:y+h, x:x+w]
            face_arr = cv2.resize(face, self.target_size)
            return face_arr
        except Exception as e:
            logging.error(f"Face detection failed for {filename}: {e}")
            return None

    def load_faces(self, dir):
        FACES = []
        for im_name in os.listdir(dir):
            path = os.path.join(dir, im_name)
            single_face = self.extract_faces(path)
            if single_face is not None:
                FACES.append(single_face)
        return FACES
    
    def load_classes(self):
        for sub_dir in os.listdir(self.directory):
            path = os.path.join(self.directory, sub_dir)
            FACES = self.load_faces(path)
            labels = [sub_dir for _ in range(len(FACES))]
            logging.info(f"Loaded {len(labels)} faces for class {sub_dir}")
            self.X.extend(FACES)
            self.Y.extend(labels)
        return np.asarray(self.X), np.asarray(self.Y)

    def plot_all_images(self):
        if not self.X:
            self.load_classes()
        
        num_images = len(self.X)
        cols = 5 
        rows = num_images // cols if num_images % cols == 0 else num_images // cols + 1
        
        fig, ax = plt.subplots(rows, cols, figsize=(15, 15))
        ax = ax.ravel()  
        
        for i in range(num_images):
            ax[i].imshow(self.X[i])
            ax[i].set_title(self.Y[i])
            ax[i].axis('off')  
            
        for i in range(num_images, rows * cols):
            ax[i].axis('off')
        
        plt.tight_layout()
        plt.show()
```

```python
# Initialize FACELOADING and load dataset
faceloading = FACELOADING(r"C:\Users\boddu\Desktop\Domain Project\Face Recognition\dataset")
X, Y = faceloading.load_classes()
```

```python
# Embedding using FaceNet
embedder = FaceNet()

def get_embedding(face_img):
    face_img = face_img.astype('float32')
    face_img = np.expand_dims(face_img, axis=0)
    yhat = embedder.embeddings(face_img)
    return yhat[0]
```

```python
EMBEDDED_X = np.asarray([get_embedding(img) for img in X])
```

```python
# Save the embeddings
np.savez_compressed('faces_embeddings.npz', EMBEDDED_X, Y)
```

```python
# Encode labels
encoder = LabelEncoder()
encoder.fit(Y)
Y_encoded = encoder.transform(Y)
```

```python
# Split data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(EMBEDDED_X, Y_encoded, shuffle=True, random_state=17)
```

```python
# SVM model with GridSearchCV for hyperparameter tuning
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}
```

```python
svc = SVC(probability=True)
model = GridSearchCV(svc, param_grid, cv=5, scoring='accuracy', verbose=2)
model.fit(X_train, Y_train)
```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score

print("Train Accuracy:", accuracy_score(Y_train, model.predict(X_train)) * 100 ,"%")
print("Test Accuracy:", accuracy_score(Y_test, model.predict(X_test)) * 100 ,"%")
```

```python
# Save the model and the label encoder together
with open('svm_model_with_encoder.pkl', 'wb') as f:
    pickle.dump({'model': model, 'encoder': encoder}, f)
```
